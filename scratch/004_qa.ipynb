```python
import pendulum

import pandas as pd
```

```python
df = pd.read_csv('cleaned_weather_data.csv')
```

```python
df = df.drop_duplicates(['utc_datetime','local_datetime', 'gems_plant_id'],keep='first')
```

```python
df = df.assign(
    utc_datetime=pd.to_datetime(df.utc_datetime),
    local_datetime=pd.to_datetime(df.local_datetime)
)
```

```python
df = df.drop(['lat', 'lon','latitude','longitude', 'plant_capacity_mw_fuel_coal'],axis=1)
```

```python
df = df[list(df.columns[-3:])+list(df.columns[:-3])]
df
```

```python
project_id = 'emissions-monitoring'
df.to_gbq('manual.wwo_weather_data',project_id=project_id)
```
